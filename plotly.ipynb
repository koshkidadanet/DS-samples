```python
import plotly

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import plotly.graph_objs as go
import plotly.express as px

from plotly.subplots import make_subplots
```

## Вывод графиков с подписями снизу

```python
x = np.arange(0, 5, 0.1)
def f(x):
    return x**2
x = np.arange(0, 5, 0.1)
y = f(x)

# создаем пустой график
fig = go.Figure()
# создадим объект точечного графика
## в названиях можно использовать html
print(go.Scatter(x=x,y=y, name='x<sup>2</sup>'))
# для добавления чего либо на график используется add_trace
## можно добавлять сколько угодно объектов
fig.add_trace(go.Scatter(x=x, y=y, name='x<sup>2</sup>'))
fig.add_trace(go.Scatter(x=x, y=y/2,name='x<sup>2</sup>/2'))
# чтобы легенда отображалась снизу в методе update_loyout меняем атрибут legend_orientation на 'h'
fig.update_layout(legend_orientation='h')
fig.show()
```

## Удаляем оступы, легенда, названия, графики

```python
x = np.arange(0, 5, 0.1)
def f(x):
    return x**2
x = np.arange(0, 5, 0.1)
y = f(x)

fig = go.Figure()

# параметр mode объекта Scatter задает вид графиков
fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='x<sup>2</sup>'))
fig.add_trace(go.Scatter(x=x, y=y/2,mode='lines+markers', name='x<sup>2</sup>/2'))
# margin меняет отступы
# c title начинаются параметры заголовка
# hovermode - как ведет себя график при наведении
# visible - шрафик виден при нажатии на легенду
fig.update_layout(
    legend_orientation='h', 
    margin= dict(l=0,r=0,t=30,b=0),
    legend = dict(x=0.5,xanchor="center"),
    title="two graphs", 
    title_x=0.5,
    hovermode='x unified',
    visible='legendonly'
)
# метод update_traces для изменения того, что находится на графике
fig.update_traces(
    hoverinfo='all',
    hovertemplate='Аргумент: %{x}<br>Функция: %{y}',
)
fig.show()
```

## Гибкое изменение точек с помощью marker

```python
x = np.arange(0, 5, 0.1)
def f(x):
    return x**2
x = np.arange(0, 5, 0.1)
y = f(x)

fig = go.Figure()

# параметр mode объекта Scatter задает вид графиков
fig.add_trace(
    go.Scatter(
        x=x, 
        y=y, 
        mode='markers', 
        name='x<sup>2</sup>',
        marker=dict(size=10),
        visible='legendonly'
    )
)
fig.add_trace(go.Scatter(x=x, y=y/2,mode='lines+markers', name='x<sup>2</sup>/2'))
# margin меняет отступы
# c title начинаются параметры заголовка
# hovermode - как ведет себя график при наведении
fig.update_layout(
    legend_orientation='h', 
    margin= dict(l=0,r=0,t=30,b=0),
    legend = dict(x=0.5,xanchor="center"),
    title="two graphs", 
    title_x=0.5,
    hovermode='x unified'
)
# метод update_traces для изменения того, что находится на графике
fig.update_traces(
    hoverinfo='all',
    hovertemplate='Аргумент: %{x}<br>Функция: %{y}',
)
# ограничить оси x,y
fig.update_yaxes(range=[0, 20])
fig.update_xaxes(range=[1, 5])
fig.show()
```

## Новые графики

```python
def f(x):
    return x**2
def h(x):
    return np.sin(x)
def k(x):
    return np.cos(x)

x = np.arange(0, 5, 0.1)
y = f(x)

# создаем фигуру с несколькими графиками (1 строка, 2 столбца)
# задаем имена каждому графику, задаем ширину столбцов
fig = make_subplots(
    rows=1, cols=2, 
    column_widths=[1, 2],
    subplot_titles=("Арифметика", "Тригонометрия")
)

# при add_trace теперь нужно указвать строку и столбец фигуры, где надо разместить элемент 
fig.add_trace(
    go.Scatter(
        x=x, 
        y=y, 
        mode='markers', 
        name='x<sup>2</sup>',
        marker=dict(size=10),
        visible='legendonly'
    ),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=x, y=y/2,mode='lines+markers', name='x<sup>2</sup>/2'),
    row=1, col=1

)
fig.add_trace(
    go.Scatter(
        x=x,
        y=h(x),
        mode='lines',
        name='sin(x)',
        visible='legendonly',
    ),
    row=1, col=2
)
fig.add_trace(
    go.Scatter(
        x=x,
        y=k(x),
        mode='lines',
        name='cos(x)',
        visible='legendonly',
    ),
    row=1, col=2
)

fig.update_layout(
    legend_orientation='h', 
    margin= dict(l=0,r=0,t=30,b=0),
    legend = dict(x=0.5,xanchor="center"),
    title="<b>2 графика</b>", 
    title_x=0.5,
    hovermode='x unified'
)

fig.update_traces(
    hoverinfo='all',
    hovertemplate='Аргумент: %{x}<br>Функция: %{y}',
)
# для ограничения осей также надо задавать row и col
fig.update_yaxes(range=[0, 20], row=1, col=1)
fig.update_xaxes(range=[1, 5],row=1, col=1)
fig.update_xaxes(title='Ось X графика 1', col=1, row=1)
fig.update_xaxes(title='Ось X графика 2', col=2, row=1)
fig.update_yaxes(title='Ось Y графика 1', col=1, row=1)
fig.update_yaxes(title='Ось Y графика 2', col=2, row=1)
fig.show()
```

## Занимаем графиком несколько ячеек, меняем размер вывода, 3d

```python
def f(x):
    return x**2
def h(x):
    return np.sin(x)
def k(x):
    return np.cos(x)

x = np.arange(0, 5, 0.1)

# в атрибуте specs настраиваем матрицу
# rowspan colspan - настройка, которая говрит, сколько строк должна занять ячейка или колонок
# None используется для удаления ячейки
# type используется для обозначения типа графика
fig = make_subplots(
    rows=2, cols=2, 
    specs=[
        [{"type": "scatter3d","rowspan": 2},{}],
        [None, {}]
    ],
    subplot_titles=("3д", "Тригонометрия", "Арифметика")
)

fig.add_trace(
    go.Scatter3d(
        x=x, y=f(x), z=x, mode='lines+markers',name='3d',
        marker=dict(size=4), line=dict(width=8)
    ),
    row=1,col=1
)
fig.add_trace(
    go.Scatter(
        x=x, 
        y=f(x), 
        mode='markers', 
        name='x<sup>2</sup>',
        marker=dict(size=10),
        visible='legendonly'
    ),
    row=2, col=2
)
fig.add_trace(
    go.Scatter(x=x, y=f(x)/2,mode='lines+markers', name='x<sup>2</sup>/2'),
    row=2, col=2

)
fig.add_trace(
    go.Scatter(
        x=x,
        y=h(x),
        mode='lines',
        name='sin(x)',
        #visible='legendonly',
    ),
    row=1, col=2
)
fig.add_trace(
    go.Scatter(
        x=x,
        y=k(x),
        mode='lines',
        name='cos(x)',
        #visible='legendonly',
    ),
    row=1, col=2
)

# с помощью height и width задаем размеры окна
fig.update_layout(
    legend_orientation='h', 
    height=500,
    width=1000,
    margin= dict(l=0,r=0,t=30,b=0),
    legend = dict(x=0.5,xanchor="center"),
    title="<b>3 графика</b>", 
    title_x=0.5,
    hovermode='x unified'
)

fig.update_traces(
    hoverinfo='all',
    hovertemplate='Аргумент: %{x}<br>Функция: %{y}',
)

fig.update_yaxes(range=[0, 20], row=1, col=1)
fig.update_xaxes(range=[1, 5],row=1, col=1)
fig.update_xaxes(title='Ось X графика 1', col=1, row=1)
fig.update_xaxes(title='Ось X графика 2', col=2, row=1)
fig.update_yaxes(title='Ось Y графика 1', col=2, row=2)
fig.update_xaxes(title='Ось X графика 2', col=2, row=2)
fig.show()
```
